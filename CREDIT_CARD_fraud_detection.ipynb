```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
data = pd.read_csv('/content/fraudTest.csv')

# Select relevant features and target
features = ['amt', 'city_pop', 'lat', 'long', 'merch_lat', 'merch_long', 'category', 'gender', 'state']  # Example features
X = data[features]
y = data['is_fraud']

# Convert categorical variables to numerical
label_encoders = {}
for column in ['category', 'gender', 'state']:
    le = LabelEncoder()
    X[column] = le.fit_transform(X[column])
    label_encoders[column] = le

# Optionally, scale the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))

# Predict on a new transaction
new_transaction = X_test[0].reshape(1, -1)  # Replace with your own data
prediction = model.predict(new_transaction)
print("Fraudulent" if prediction[0] == 1 else "Legitimate")
```
